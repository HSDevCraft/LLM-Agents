```python
from dotenv import load_dotenv
import os
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv('GROQ_API_KEY')
```

```python

from langchain_groq import ChatGroq

GROQ_LLM = ChatGroq(
            # api_key=os.getenv("GROQ_API_KEY"),
            model="llama3-70b-8192"
        )
```

```python

from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools.ddg_search.tool import DuckDuckGoSearchResults
search_tool = DuckDuckGoSearchRun()
# search_tool = DuckDuckGoSearchResults()
     
```

```python
search_tool.run("Tell me about GenAI?")
```

```python
from crewai import Agent, Task, Crew, Process
```

```python
# Define agent along with your goals
researcher = Agent(
    role="Research & Developement Engineer",
    goal="Uncover cutting-edge developments in AI and data science",
    backstory="""As an R&D engineer in GenAI, you pioneer AI-driven solutions for genetic analysis.
                 Crafting algorithms and software, you decode genomic data to predict traits and diseases.
                 Collaborating across disciplines, you accelerate research in personalized medicine and agriculture, shaping the future of genetic understanding and application.""",
    llm=GROQ_LLM,
    verbose=True,
    allow_delegation=False
)

senior_documentation_specialist = Agent(
    role='Senior Documentation Specialist',
    goal="Provide detailed documentation on the latest AI and data science innovations",
    backstory="""You are a renowned Senior System Design Engineer, known for your insightful and engaging outputs.
                 You transform complex concepts into compelling narratives & deliver to user in formatted way.""",
    llm=GROQ_LLM,
    verbose=True,
    allow_delegation=True
)
```

```python
# Create task for your agents
task1 = Task(
    description="""Perform an exhaustive examination of the Multi AI Agent Systems with crewAI short course launched by DeepLearning.AI.
    Identify pivotal trends, groundbreaking technologies, and their potential ramifications across industries.""",
    expected_output="Full analysis report in bullet points",
    agent=researcher,
    tools=[search_tool],
)

task2 = Task(
    description="""Using the above report, provide detailed documentation on the latest AI and data science innovations.""",
    expected_output="Write deatiled linkedin post on Multi AI Agent Systems with crewAI short course launched by DeepLearning.AI in bullet points & detailed manner for posting it in on LinkedIn & share the learnings with community.",
    agent=senior_documentation_specialist,
    tool=[search_tool]
)
```

```python

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[researcher, senior_documentation_specialist],
  tasks=[task1, task2],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)
```

```python

# Get your crew to work!
result = crew.kickoff()

print("###########")
print(result)
     
```
